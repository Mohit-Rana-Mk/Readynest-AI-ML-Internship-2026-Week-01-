### Load Dataset
Read the Superstore dataset into a Pandas DataFrame for further analysis.

```python
df = pd.read_csv('../dataset/sample_-_superstore.csv')

df.head()
```

```python
df_clean = df.copy()
```

## Data Cleaning
Process Missing Data and Duplicates

Handle missing values, remove duplicate records, convert date columns, and standardize column names.

```python
missing = df_clean.isnull().sum()

missing_percent = (
    missing / len(df_clean)
) * 100

pd.DataFrame({
    "Missing Values": missing,
    "Percentage": missing_percent
}).sort_values(
    by="Missing Values",
    ascending=False
)
```

```python
df_clean['Category'] = df_clean['Category'].fillna(
    df_clean['Category'].mode()[0]
)
```

```python
df_clean.duplicated().sum()
```

```python
df_clean = df_clean.drop_duplicates()
```

```python
df_clean.duplicated().sum()
```

```python
df_clean.dtypes
```

```python
df_clean['Order Date'] = pd.to_datetime(
    df_clean['Order Date']
)

df_clean['Ship Date'] = pd.to_datetime(
    df_clean['Ship Date']
)
```

```python
df_clean.info()
```

```python
df_clean.columns = (
    df_clean.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
```

## Data Visualization
Compare Categories, Segments, Regions and Trends

```python
category_profit = (
    df_clean.groupby('category')['profit']
    .sum()
    .sort_values(ascending=False)
)

category_profit.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title('Profit by Category')
plt.ylabel('Profit')
plt.show()
```

```python
region_sales = (
    df_clean.groupby('region')['sales']
    .sum()
    .sort_values(ascending=False)
)

region_sales.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title('Sales by Region')
plt.ylabel('Sales')
plt.show()
```

```python
region_profit = (
    df_clean.groupby('region')['profit']
    .sum()
    .sort_values(ascending=False)
)

region_profit.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title('Profit by Region')
plt.ylabel('Profit')
plt.show()
```

```python
segment_sales = (
    df_clean.groupby('segment')['sales']
    .sum()
    .sort_values(ascending=False)
)

segment_sales.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title('Sales by Customer Segment')
plt.ylabel('Sales')
plt.show()
```
